```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

# Initialize kernels correctly
check_kernal_01 = np.zeros((3, 3)) / 9
check_kernal_02 = np.zeros((3, 3)) / 9

check_kernal_03 = np.array([[1, 1, 0],
                            [1, -2, 1],
                            [0, 0, -2]])

# Display kernel 1
plt.imshow(check_kernal_01, cmap='gray')
plt.colorbar()
plt.title('Kernel with zeros visualization')  # Corrected spelling of 'title'
plt.show()

# Display kernel 2
plt.imshow(check_kernal_02, cmap='gray')
plt.colorbar()
plt.title('Kernel with zeros visualization')  # Corrected spelling of 'title'
plt.show()

# Display kernel 3
plt.imshow(check_kernal_03, cmap='gray')
plt.colorbar()
plt.title('Custom kernel visualization')  # Corrected spelling of 'title'
plt.show()
```
